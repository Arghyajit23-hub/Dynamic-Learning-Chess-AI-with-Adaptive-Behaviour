```python
import pandas as pd

df = pd.read_csv('PlayerVsOpponent.csv')
print(f"--- Stats for PlayerVsOpponent.csv ({len(df)} games) ---")

c, total = df['Result'].value_counts(), len(df)
print(f"Win: {c.get(1.0,0)/total*100:.2f}% | Draw: {c.get(0.0,0)/total*100:.2f}% | Loss: {c.get(-1.0,0)/total*100:.2f}%")
print("-"*30)

d = df[(df['Result'] != 0.0) & (df['Moves'] != 120)]
print(f"Avg moves (decisive games): {d['Moves'].mean():.2f} ({len(d)}/{total})" if len(d) else "No decisive games.")
```
